```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale=1/255, validation_split=0.2)

train = datagen.flow_from_directory(
    "artifacts/data_ingestion/data_kidney",
    target_size=(224,224),
    subset="training",
    class_mode="categorical"
)

print(train.class_indices)
```

```python
# ---- Imports ----
from cnnClassifier.config.configuration import ConfigurationManager
from cnnClassifier.components.training import Training
from cnnClassifier.components.prepare_callbacks import PrepareCallback

# ---- Load configs ----
config = ConfigurationManager()
training_config = config.get_training_config()
callback_config = config.get_prepare_callback_config()

# ---- Prepare callbacks ----
callbacks = PrepareCallback(callback_config).get_tb_ckpt_callbacks()

# ---- Start Training ----
trainer = Training(training_config)
trainer.get_base_model()       # loads updated base model
trainer.train(callback_list=callbacks)
```
