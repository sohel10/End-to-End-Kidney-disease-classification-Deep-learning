```python
model = tf.keras.models.load_model("artifacts/training/model1.h5")
print("Model loaded successfully!")
```

```python
datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    validation_split=0.30
)

valid_gen = datagen.flow_from_directory(
    "artifacts/data_ingestion/data_kidney",
    subset="validation",
    shuffle=False,
    target_size=(224,224),
    batch_size=32
)
```

```python
loss, acc = model.evaluate(valid_gen)
print("Loss:", loss)
print("Accuracy:", acc)
```
